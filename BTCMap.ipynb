# 获取比特币成交日期，价格并绘制成图表

```python
# coding:utf-8
import matplotlib.pyplot as plt
import requests
import json
import time
```

```python
url = "https://api.btctrade.com/api/trades?coin=btc"

response = requests.get(url)
res_con = response.content.decode('utf-8')
res_dict = json.JSONDecoder().decode(res_con)
```

```python
data_btc = []
dates = []
for datas in res_dict:
    data_btc.append(datas['price'])
    
for date_p in res_dict:
    dates.append(time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(int(date_p['date']))))
```

```python
%matplotlib inline
fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(data_btc)
plt.title("BTC",fontsize=24)
plt.xlabel("Date",fontsize=14)
fig.autofmt_xdate()
plt.xticks((0,10,20,30,40,50),dates)
plt.ylabel("Price",fontsize=14)
# plt.ylabel("Price",fontsize=14)
# plt.tick_params(labelsize=14)
plt.show()
```
